```python
import pandas
import matplotlib.pyplot as plt
import seaborn

data = pandas.read_csv("data/Clean_Quasar_Data.csv")

n_samples = 100

# Takes a random subset of each class
subset_0 = data[data['BAL_PROB'] == "Non-BAL"].sample(n=n_samples, random_state=42)
subset_1 = data[data['BAL_PROB'] == "BAL"].sample(n=n_samples, random_state=42)

# Combines the subsets into a single DataFrame
subset_data = pandas.concat([subset_0, subset_1])

# Separates data by BAL_PROB
bal_prob_0 = subset_data[subset_data['BAL_PROB'] == "Non-BAL"][['W1_FLUX_LUMINOSITY', 'PSFFLUX_0_LUMINOSITY', 'W2_FLUX_LUMINOSITY', 'NUV_LUMINOSITY', 'PSFFLUX_1_LUMINOSITY', 'PSFFLUX_4_LUMINOSITY', 'GAIA_BP_FLUX_SNR_LUMINOSITY', 'PSFFLUX_2_LUMINOSITY']]
bal_prob_1 = subset_data[subset_data['BAL_PROB'] == "BAL"][['W1_FLUX_LUMINOSITY', 'PSFFLUX_0_LUMINOSITY', 'W2_FLUX_LUMINOSITY', 'NUV_LUMINOSITY', 'PSFFLUX_1_LUMINOSITY', 'PSFFLUX_4_LUMINOSITY', 'GAIA_BP_FLUX_SNR_LUMINOSITY', 'PSFFLUX_2_LUMINOSITY']]

# Calculates correlations
corr_0 = bal_prob_0.corr()
corr_1 = bal_prob_1.corr()
```

```python
bal_prob_counts = data['BAL_PROB'].value_counts()

bal_prob_table = pandas.DataFrame({'BAL_PROB': bal_prob_counts.index, 'Frequency': bal_prob_counts.values})

print(bal_prob_table)
```

```python
import numpy

# Create a mask for the upper triangle
mask = numpy.triu(numpy.ones_like(corr_0, dtype=bool))

# Set up the matplotlib figure with side-by-side subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# Plot the heatmap for BAL_PROB = 0
seaborn.heatmap(corr_0, mask=mask, annot=True, cmap='Blues', square=True, ax=ax1, cbar_kws={'shrink': .8})
ax1.set_title('Correlation Matrix for BAL_PROB = 0')

# Plot the heatmap for BAL_PROB = 1
seaborn.heatmap(corr_1, mask=mask, annot=True, cmap='Blues', square=True, ax=ax2, cbar_kws={'shrink': .8})
ax2.set_title('Correlation Matrix for BAL_PROB = 1')

plt.tight_layout()
plt.show()
```
